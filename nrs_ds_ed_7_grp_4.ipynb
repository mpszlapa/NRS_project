Here, a function has been created to clone the Git repository into the current workspace of the Jupyter notebook. The details of this function are explained in its docstring, a widely used method for documentation in Python.

```python
def clone_git_repo(repo_url, access_token, clone_dir):
  '''
    Clones a Git repository from a given URL into a specified directory.

    This function deletes any existing directory at the 'clone_dir' path before
    cloning.If cloning is successful, the function returns a success message,
    otherwise, it returns a failure message.

    Args:
    repo_url (str): The URL of the Git repository to clone.
    access_token (str): Access token for authenticating with the Git repository.
    clone_dir (str): The local directory path where the repository will be
                     cloned.

    Returns:
    str: A message indicating whether the repository was successfully cloned or
         if cloning failed.

    Raises:
    subprocess.CalledProcessError: If the 'git clone' command fails.
  '''

  # Delete an existing directory if it exists
  if os.path.exists(clone_dir):
    shutil.rmtree(clone_dir)

  # Compose the URL for accessing the repository
  url_with_token = repo_url.replace('https://',
                                    f'https://x-access-token:{access_token}@')

  try:
      # Clone the repository into the specified directory
      subprocess.run(['git', 'clone', url_with_token, clone_dir], check = True)
      return 'Repository cloned successfully.'
  except subprocess.CalledProcessError:
      return 'Failed to clone repository.'


clone_git_repo(
    repo_url = repo,
    access_token = token,
    clone_dir = path
)
```

## 3.1 Transform and restructure the training data within Pandas DataFrame <a class="anchor" id="section_3_1"></a>

To start with, we have developed a function that addresses the column-related issues in the dataset. The outcome is a dataset with the columns correctly separated. All details of this function are well-described using comments and docstring.

```python
def transform_train_data(df_in):
  '''
    Transforms the Pandas DataFrame of the training data and restructures
    it in order to rectify the column issues.

    Args:
    df_in (pandas.DataFrame): The input DataFrame containing the training data.

    Returns:
    pandas.DataFrame: The transformed DataFrame, ready for further processing.

  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Concatenate rows of all columns to a comma-separated string
  df = df.apply(lambda row: ','.join(row.values.astype(str)), axis = 1)

  # Rectify various representations of a dot by replacements
  df = df.replace('&#x2e,', '.', regex = True) \
                    .replace('U\+0002E', '.', regex = True) \
                    .replace('&#46,', '.', regex = True) \
                    .replace('&period,', '.', regex = True)

  # Split the string into columns by commas (commas in quotes are neglected)
  df = df.str.split(',(?=(?:[^\"]*\"[^\"]*\")*[^\"]*$)', expand = True)

  # Set the first row as column header
  df.columns = df.iloc[0]

  # Drop the first row after setting it as column header
  df.drop(df.index[0], axis = 0, inplace = True)

  # Drop the first column and the two last ones
  df.drop(df.iloc[:, [0, 20, 21]].columns, axis = 1, inplace = True)

  return df


df_train_transformed = transform_train_data(
    df_in = df_train
)

df_train_transformed.head()
```

## 3.2 Clean the training data and perform data type conversions <a class="anchor" id="section_3_2"></a>

In this section, we have created a function that performs fundamental data cleaning steps and handles data type conversions. Similarly to what was mentioned earlier, this function is well-documented using comments and docstring.

The column 'average_investment' does not contain any meaningful information. There are only two values for this column: '$%â‚¬_53f""e' or nan. Even if the other columns may allow the computation of this column, without additional details or a specific formula, we can only make some guesses. After reaching a collective decision with the entire team, we have concluded that making guesses is not an appropriate approach for this situation. Therefore, we have decided to exclude this column from our dataset.

```python
def clean_train_data(df_in):
  '''
    Cleans the data of the previously transformed training DataFrame and
    performs data type conversions to ensure correct representation of data.

    Args:
    df_in (pandas.DataFrame): The input DataFrame containing the transformed
                              training data.

    Returns:
    pandas.DataFrame: The cleaned DataFrame, ready for further processing.

  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Clean the 'insurance_class' column
  df['insurance_class'] = df['insurance_class'].str \
    .replace('\"', '', regex = True) \
    .replace('and', '', regex = True) \
    .replace(',\s+', ',', regex = True) \
    .replace('\s', '_', regex = True) \
    .replace('1st', 'First', regex = True) \
    .replace('2nd', 'Second', regex = True) \
    .replace('', pd.NA, regex = True)

  # Clean the 'number_of_operations' column
  df['number_of_operations'] = df['number_of_operations'].str \
    .replace('\$', '', regex = True) \
    .replace('#', '', regex = True)

  # Drop the 'average_investment' column as it lacks of meaningful information
  df.drop('average_investment', axis = 1, inplace = True)

  # Map the 'month' column to integer values
  month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
  }
  df['month'] = df['month'].replace(month_mapping)

  # Replace empty strings of the 'account_movements_index' column with pd.NA
  df['account_movements_index'] = df['account_movements_index'] \
    .replace('', pd.NA, regex = True)

  # Replace empty strings of the 'balance' column with pd.NA
  df['balance'] = df['balance'].replace('', pd.NA, regex = True)

  # Replace -999 of the 'interest_rate' column with pd.NA
  df['interest_rate'] = df['interest_rate'].replace('-999', pd.NA, regex = True)

  # Replace -999 of the 'late_payments' column with pd.NA
  df['late_payments'] = df['late_payments'].replace('-999', pd.NA, regex = True)

  # Replace (Unknown, Yes, No) of the 'deposit' column with (pd.NA, 1, 0)
  df['deposit'] = df['deposit'].replace('Unknown', pd.NA, regex = True) \
                               .replace('Yes', 1, regex = True) \
                               .replace('No', 0, regex = True)

  # Convert the specified columns to Pandas nullable integer type ('Int64')
  columns_to_int = ['id', 'age', 'bank_accounts', 'insurance_products',
                    'current_debt', 'month', 'paymant_delay', 'late_payments',
                    'number_of_operations', 'account_movements_index',
                    'deposit', 'score']
  df[columns_to_int] = df[columns_to_int] \
    .apply(lambda col: col.astype('Float64').astype('Int64'))

  # Convert the specified columns to Pandas nullable float type ('Float64')
  columns_to_float = ['income_year', 'wage_month', 'interest_rate', 'balance']
  df[columns_to_float] = df[columns_to_float] \
    .apply(lambda col: col.astype('Float64'))

  # Convert the specified columns to Pandas nullable string type
  columns_to_str = ['job', 'insurance_class']
  df[columns_to_str] = df[columns_to_str] \
    .apply(lambda col: col.astype('string'))

  # Rename column 'paymant_delay' -> 'payment_delay' to fix spelling error
  df.rename(columns={'paymant_delay': 'payment_delay'}, inplace = True)

  return df


df_train_cleaned = clean_train_data(
    df_in = df_train_transformed
)

df_train_cleaned.head()
```

## 3.3 Decompose the list column 'insurance_class' into multiple columns <a class="anchor" id="section_3_3"></a>

We have developed a function that enables the decomposition of a single list column into multiple columns containing binary flags. The inner working of this function is described using comments and docstring.

Although this function is designed to be generic and reusable, in our particular use case, it is solely required for the 'insurance_class' column.

```python
def decompose_list_column(df_in, list_column):
  '''
    Decomposes a column in a DataFrame that contains lists into
    multiple columns.

    Args:
    df_in (pd.DataFrame): The input DataFrame.
    list_column (str): The name of the column containing list values.

    Returns:
    pd.DataFrame: A DataFrame with new columns, each representing a unique
    element from the list column, indicating its presence with a binary flag
    (1 for present, 0 for not present).
  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Extract unique elements from the lists in the specified column
  list_attrs = set(','.join(df[df[list_column].notnull()][list_column] \
                            .to_list()).split(','))

  # Create new columns for each unique element in the list column
  for elem in list_attrs:

    # Convert the new column name to lowercase
    new_column = list_column + '_' + elem.lower()

    # Determine the index of the list column
    col_index = df.columns.get_loc(list_column)

    # Insert the new column after list column
    df.insert(col_index + 1, new_column, pd.NA)

    # Create a new column for each element
    # Set to 1 if the element is present, otherwise 0
    df[new_column] = np.where(df[list_column] \
                              .str.contains(elem, na = False), 1, 0)

    # Convert data type to Pandas nullable integer type ('Int64')
    df[new_column] = df[new_column].astype('Int64')

  # Drop the list column
  df.drop(list_column, axis = 1, inplace = True)

  return df


df_train_decomposed = decompose_list_column(
    df_in = df_train_cleaned,
    list_column = 'insurance_class'
)

df_train_decomposed.head()
```

Alternative class to decompose a categorical column into binary columns

```python
class CustomOneHotEncoder():
  _separator = None
  _column = None
  _classes = pd.Index([])

  def fit(self, s, separator = "|"):
    '''
    Uses the method get_dummies to identify the list of distinct values in the
    input series. The method can handle multiple values and receive in input a
    separator for the split. The list is saved in the variable _classes

    Args:
    s (Series): The input column to be splitted into binary columns.
    separator (str): separator used to split string that contains
    multiple values

    Returns:
    None.

    '''
    #set interal separator anc column name
    self._separator = separator
    self._column = s.name
    #get dummies and save into variable
    # s=s.fillna("")
    dummies_train = s.fillna("").apply(lambda x : x.replace("\"",""
    ).lower()).str.get_dummies(self._separator)
    self._classes = dummies_train.columns

  def transform(self, s):
    '''
    Uses the method get_dummies to identify the list of distinct values in
    the input series. Iterates the variable _classes to generate the columns
    identified during the fit. New values do not generate additional columns.

    Args:
    s (Series): The input column to be splitted into binary columns.

    Returns:
    panda.Dataframe: dataframe containg the binary column generated by the split

    '''
    #get dummies
    dummies = s.fillna("").apply(lambda x : x.replace("\"",""
    ).lower()).str.get_dummies(self._separator)
    #initialize dataframe
    idf = pd.DataFrame([],index=s.index)
    #ignore new dummy columns that were not detected in during the training
    for c in self._classes:
      if c in dummies.columns:
        #add column if exist
        idf[c]=pd.Series(dummies[c])
      else:
        #add 0s id the dummy column doesn't exist in the test set
        idf[c]=pd.Series(0,index = s.index)
    #rename columns
    idf.rename(columns=lambda n: f"{self._column}_{n}", inplace=True)
    #return
    return idf
```

```python
def decompose_categorical_column(df_in, column_name, encoder_path, is_train,
                                 drop=True):
  '''
  Create an encoder and call the transform method.
  If the flag is_train is equal to true then the fit method is called first and
  the encoder is saved into a file.
  Otherwise the encoder is loaded from a file.

  Args:
  df_in (datafram): input dataframe.
  column_name (str): Name of the original colum
  is_train (bool): True if the method is executed on the training set,
                   False otherwise
  separator (str): separator passed to the encoder

  Returns:
  None.

  '''
  df = df_in.copy()

  encoder_file_name = f'{encoder_path}/{column_name}_encoder.pkl'

  enc = CustomOneHotEncoder()
  if is_train:
    enc.fit(df[column_name], separator=",") #", and "
    pickle.dump(enc, open(encoder_file_name, 'wb'))
  else:
    enc = pickle.load(open(encoder_file_name, 'rb'))

  df_new = enc.transform(df[column_name])
  df = pd.concat([df,df_new],axis=1)
  if drop:
    df.drop([column_name],axis=1,inplace=True)
  return df
```

## 4.1 Tailored functions for exploratory data analysis <a class="anchor" id="section_4_1"></a>

We developed two distinct functions: one for analyzing numeric variables and another for categorical/ordinal variables. ### 4.1.1 Function for analyzing numeric columns <a class="anchor" id="section_4_1_1"></a>

This function is intended to analyze numeric columns. Therefore, the following steps are performed:
* compute descriptive statistics (e.g. mean/median, standard deviation, etc.)
* calculate the amount of missing values
* detect outliers based on the interquartile range
* compute Pearson correlation coefficients among other numeric columns
* plot histogram to visualize the distribution
* create boxplots based on the target variable

```python
def analyze_numeric_column(df, col_name, numeric_cols, hist_bins,
                           corr_sig_level = 0.3, boxplot_min = None):
    '''
      Analyzes a numeric column in a DataFrame.

      Args:
      df (pd.DataFrame): The DataFrame containing the data.
      col_name (str): The name of the column to analyze.
      numeric_cols (list): List of numeric columns for correlation analysis.
      hist_bins (int): Number of bins for the histogram.
      corr_sig_level (float): The significance level for correlation.
      boxplot_min (float or None): Minimum value for boxplot filtering.

      This function performs the following analyses on the specified column:
      - Prints descriptive statistics.
      - Calculates and prints the percentage of missing values.
      - Detects and prints information about outliers.
      - Calculates and prints Pearson correlation coefficients with other
        numeric columns.
      - Generates a histogram.
      - Generates boxplots.
    '''

    # Descriptive statistics
    print(f'\nDescriptive statistics of {col_name}:')
    print(df[col_name].describe().to_string())

    # Missing values analysis
    print(f'\nMissing values of {col_name}:')
    missing_values_count = df[col_name].isna().sum()
    missing_values_percentage = round(missing_values_count / df[col_name] \
                                      .count() * 100, 2)
    print(f'{missing_values_count} ({missing_values_percentage}%)')

    # Outlier detection
    print(f'\nOutlier detection of {col_name}:')
    first_quartile = df[col_name].quantile(0.25)
    third_quartile = df[col_name].quantile(0.75)
    interquartile_range = third_quartile - first_quartile
    lower_bound = first_quartile - 1.5 * interquartile_range
    upper_bound = third_quartile + 1.5 * interquartile_range

    lower_outliers = df[df[col_name] < lower_bound]
    upper_outliers = df[df[col_name] > upper_bound]
    lower_outliers_count = lower_outliers[col_name].count()
    upper_outliers_count = upper_outliers[col_name].count()
    lower_outliers_percentage = round(lower_outliers_count / df[col_name] \
                                      .count() * 100, 2)
    upper_outliers_percentage = round(upper_outliers_count / df[col_name] \
                                      .count() * 100, 2)

    print(f'Lower bound (Q1 - 1.5 * IQR): {lower_bound} -> \
      {lower_outliers_count} ({lower_outliers_percentage}%) outliers detected')
    print(f'Upper bound (Q3 + 1.5 * IQR): {upper_bound} -> \
      {upper_outliers_count} ({upper_outliers_percentage}%) outliers detected')

    # Correlation analysis
    print(f'\nPearson correlation coefficients of {col_name} \
    with significance level greater or equal to {corr_sig_level}:')
    for index, value in df[numeric_cols] \
      .corr(method = 'pearson')[col_name].items():
        if index != col_name and abs(value) >= corr_sig_level:
            print(f'{index.ljust(30)} {value}')

    # Histogram plotting
    print('\n')
    plt.hist(df[df[col_name].notna()][col_name], bins = hist_bins)
    plt.title(f'Histogram of {col_name}')
    plt.xlabel(col_name)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    # Boxplot plotting
    sns.boxplot(data = ((df[df[col_name] > boxplot_min]) if \
      boxplot_min is not None else df), x = 'score', y = col_name)
    plt.title(f'Boxplots of {col_name} according to different scores ' + \
      (f'({col_name} > {boxplot_min})' if boxplot_min is not None else ''))
    plt.show()
```

### 4.1.2 Function for analyzing categorical/ordinal columns <a class="anchor" id="section_4_1_2"></a>

This function is intended to analyze categorical/ordinal columns. Therefore, the following steps are performed:
* compute a frequency table
* calculate the mode
* compute the amount of missing values
* calculate Spearman rank correlation coefficients (only if column is ordinal)
* visualize frequency distribution using a barplit
* generates barplots based on the target variable

```python
def analyze_categorical_column(df, col_name, ordinal_cols, corr_sig_level = 0.3,
                               show_score_barplot = True,
                               sort_freq_index = False):
    '''
      Analyzes a categorical/ordinal column in a DataFrame.

      Args:
      df (pd.DataFrame): The DataFrame containing the data.
      col_name (str): The name of the column to analyze.
      ordinal_cols (list): List of ordinal columns for correlation analysis.
      corr_sig_level (float): The significance level for correlation.
      show_score_barplot (bool): Flag to indicate if barplot of relative
        frequencies according to different scores should be shown.
      sort_freq_index (bool): Flag to indicate if the frequency index should
        be sorted numerically before plotting.

      This function performs the following analyses on the specified column:
      - Prints a frequency table.
      - Calculates and prints the mode.
      - Calculates and prints the percentage of missing values.
      - If the column is ordinal, calculates and prints Spearman rank
        correlation coefficients.
      - Generates a bar plot of the frequency distribution.
      - If indicated, generates a barplot of relative frequencies according
        to different scores.
    '''

    # Frequency table
    print(f'\nFrequency table of {col_name}:')
    frequency = df[col_name].value_counts().sort_index()
    percentage = (df[col_name].value_counts(normalize = True) * 100)
    freq_table = pd.DataFrame({'frequency': frequency,
                               'percentage': percentage})
    print(f'{freq_table.to_string()}')

    # Mode calculation
    print(f'\nMode of {col_name}:')
    print(df[col_name].mode()[0])

    # Missing values analysis
    print(f'\nMissing values of {col_name}:')
    missing_values_count = df[col_name].isna().sum()
    missing_values_percentage = round(missing_values_count / \
                                      df[col_name].count() * 100, 2)
    print(f'{missing_values_count} ({missing_values_percentage}%)')

    # Correlation analysis for ordinal columns
    if col_name in ordinal_cols:
        print(f'\nSpearman rank correlation coefficients of {col_name} \
        with significance level greater or equal to {corr_sig_level}:')
        for index, value in df[ordinal_cols] \
          .corr(method = 'spearman')[col_name].items():
            if index != col_name and abs(value) >= corr_sig_level:
                print(f'{index.ljust(30)} {value}')

    # Convert the index to integer and sort by the index
    if sort_freq_index:
      frequency.index = frequency.index.astype(int)
      frequency = frequency.sort_index()

    # Frequency bar plot
    plt.figure(figsize=(10, 4))
    frequency.plot(kind='bar')
    plt.title(f'Barplot of {col_name}')
    plt.xlabel(col_name)
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)
    plt.show()

    # Score-based relative frequency bar plot
    if show_score_barplot:
        cross_tab_prop = pd.crosstab(df[col_name], df['score'], \
                                     normalize = 'index').round(4) * 100
        ax = cross_tab_prop.plot(kind = 'bar', stacked = True, \
                                 figsize = (10, 4))
        ax.yaxis.set_major_formatter(mtick.PercentFormatter())
        ax.margins(y = 0.15)
        ax.legend(loc = 'upper center', ncol = 4)
        for n, x in enumerate([*cross_tab_prop.index.values]):
            for (proportion, y_loc) in zip(cross_tab_prop.loc[x], \
                                           cross_tab_prop.loc[x].cumsum()):
                plt.text(x = n, y = (y_loc - proportion) + (proportion / 2),
                         s = f'{proportion:.0f}%',
                         color = 'black',
                         fontsize = 10)
        plt.title(f'Relative frequency distribution of {col_name} \
          according to different scores')
        plt.show()
```

### 4.1.3 Identify numeric and ordinal columns for correlation analysis <a class="anchor" id="section_4_1_3"></a>

Next, we need to identify which columns can be treated as numeric or ordinal, so that they can be appropriately utilized later for calculating correlation coefficients using Pearson and Spearman rank correlation methods.

The outcome is two lists: one with the names of numeric columns and another with the names of at least ordinal columns, including both numeric and ordinal columns.

This distinction is required for calculating correlation coefficients: Pearson's correlation is applicable only to numeric columns, whereas Spearman's correlation can be calculated for columns that are at least ordinal.

```python
# Select columns of type Float64 from the dataframe 'df_cleaned'
float64_columns = list(df_cleaned.select_dtypes(include=['Float64']).columns)

# Specify additional columns that are numeric but not necessarily of
# type Float64
additional_numeric_columns = list(df_cleaned[['age', 'bank_accounts',
                                         'number_of_operations',
                                         'insurance_products',
                                         'current_debt', 'payment_delay',
                                         'late_payments']].columns)

# Combine the columns of type Float64 with the manually specified
# numeric columns
numeric_columns = float64_columns + additional_numeric_columns
print(f'Numeric columns for Pearson correlation analysis:\n{numeric_columns}')

# Select columns that contain ordinal data
ordinal_columns = list(df_cleaned[['account_movements_index', 'month',
                                   'score']].columns)

# Combine the numeric columns with the ordinal columns to get a list of columns
# that are at least ordinal
at_least_ordinal_columns = numeric_columns + ordinal_columns
print(f'\nAt least ordinal columns for Spearman rank correlation \
  analysis:\n{at_least_ordinal_columns}')
```

## 5.10 Job: One-hot encoding of categorical values <a class="anchor" id="section_5_10"></a>

We have developed a function that enables the one-hot encoding of categorical columns each into multiple columns containing binary flags.

The inner working of this function is described using comments and docstring.

Although this function is designed to be generic and reusable, in our particular use case, it is solely required for the 'job' column.

```python
def perform_one_hot_encoding(df_in, features):
  '''
    Perform one-hot encoding on specified features of a DataFrame.

    Args:
    df_in (pandas.DataFrame): The input DataFrame on which one-hot encoding
                              will be performed.
    features (list of str): A list of column names in the DataFrame to be
                            one-hot encoded.

    Returns:
    pandas.DataFrame: A DataFrame with one-hot encoded features.
                      All column names are in lowercase.
  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Perform one-hot encoding for features
  for feature in features:
    df_cat = df[feature]
    df = pd.get_dummies(df, columns = [feature])
    df[feature] = df_cat

    # Convert the created columns to Pandas nullable integer type ('Int64')
    for column in df.columns:
      if column.startswith(f'{feature}_'):
        df[column] = df[column].astype('Int64')


  # Convert all column names to lowercase
  df.columns = [col.lower() for col in df.columns]

  return df


# Determine a list of features for one-hot encoding
features_to_encode = ['job']

df_encoded = perform_one_hot_encoding(
    df_in = df_deposit_complete,
    features = features_to_encode
)

df_encoded.head()
```

```python
def perform_standard_scaling(df_in, variables):
  '''
    Applies standard scaling to specified columns in a given DataFrame.

    Args:
    df_in (pd.DataFrame): The input DataFrame containing the variables
                          to be scaled.
    variables (list of str): A list of column names in the DataFrame
                            to be scaled.

    Returns:
    pd.DataFrame: A DataFrame with the specified variables scaled and
                  certain columns converted to specific data types
                  (Int64 or Float64).
  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Scale each variables in the provided list of variables
  for variable in variables:
      scaler = StandardScaler()
      df[variable + '_scaled'] = scaler.fit_transform(df[[variable]])

   # Convert the specified columns to Pandas nullable float type ('Float64')
  columns_to_float = [variable + '_scaled' for variable in variables]
  df[columns_to_float] = (df[columns_to_float]
    .apply(lambda col: col.astype('Float64')))

  return df


# Determine a list of variables for standard scaling
variables_to_scale = [
    'age',
    'income_year',
    'income_year_log',
    'wage_month',
    'wage_month_log',
    'bank_accounts',
    'interest_rate_complete',
    'number_of_operations',
    'balance_complete',
    'balance_cbrt',
    'insurance_products',
    'current_debt',
    'payment_delay',
    'late_payments_complete'
]

df_scaled = perform_standard_scaling(
    df_in = df_encoded,
    variables = variables_to_scale
)

df_scaled.head()
```

### 5.12.1 Factor analysis <a class="anchor" id="section_5_12_1"></a>

We run Factor Analysis to check whether can reduce the number of features. We keep scaled variables on the dataset what we use in this analysis. We skip the target variables because we would like to use the factors as independent variables in our final model.

First we check whether our dataset is factorability. Bartlett’s test of sphericity checks whether or not the observed variables intercorrelate at all using the observed correlation matrix against the identity matrix. If the p-values is lower than 0.05, it means the test is significant.

Kaiser-Meyer-Olkin (KMO) Test measures the suitability of data for factor analysis. It determines the adequacy for each observed variable and for the complete model. KMO estimates the proportion of variance among all the observed variable. If the value of KMO less than 0.6 is considered inadequate.

After we check the main criterions of the factor analysis, we have to see the value of Eigenvalues by selected variables. The number of factors can be decided on the basis of the amount of common variance the factors explain. Eigenvalues are the amount of variance the factor explains. The factors where the Eigenvalue is higher than 1, explain more variance than a single observed variable. Our approach is that we choose the number of factors with more variance.

The cumulative variance will show the cumulative sum of proportional variances of each factor. The total variance should be higher than 60%. It means the x number of factors together are able to explain at least the 60% of the total variance.

```python
def perform_factor_analysis(df_in, variables, ei_value):
  '''
    Perform factor analysis (FA) on a given DataFrame with
    specified variables.

    Args:
    df_in (pd.DataFrame): The input DataFrame.
    variables (list): A list of column names from df_in to be used for FA.
    variance (float): The fraction of total variance to retain during FA.
                      This should be a float between 0.0 and 1.0.

    Returns:
    pd.DataFrame: A DataFrame containing the original data and the
                  factor components. The factor components are named as
                  'fa_1', 'fa_2', ..., 'fa_n'.
  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Create a DataFrame with the selected variables
  df_selected = df[variables]

  # Transform the DataFrame into float type
  df_selected = df_selected.astype(float)

  # Check whether our dataset is factorability with Bartlett’s test of
  # sphericity
  chi_square_value,p_value=calculate_bartlett_sphericity(df_selected)
  print(f'Bartlett’s test of sphericity: {chi_square_value, p_value}')

  # Check Kaiser-Meyer-Olkin (KMO) test
  kmo_all,kmo_model=calculate_kmo(df_selected)
  print(f'\nKaiser-Meyer-Olkin (KMO) test: {kmo_model}')

  # Create factor analysis object and perform factor analysis
  fa = FactorAnalyzer()
  fa.fit(df_selected)

  # Check Eigenvalues
  ev, v = fa.get_eigenvalues()
  print(f'\nEigenvalues: {ev}')

  # Get the number of selected variables
  nr_of_variables = len(variables)
  print(f'\nNumber of selected variables: {nr_of_variables}')

  # Use number of factor where the Eigenvalue is higher than the value what
  # we selected. Mostly we keep number of factors where Eigenvalue higher than 1
  number_factors = len(ev[ev > ei_value])
  print(f'\nNumber of factors: {number_factors}')

  # Degree of dimensionality reduction
  fa_dim_red = nr_of_variables - number_factors
  print(f'\nDegree of dimensionality reduction by FA: {fa_dim_red}')

  # Create factor analysis object and perform factor analysis
  fa = FactorAnalyzer(n_factors=number_factors)
  fa.fit(df_selected)
  fa_components = fa.fit_transform(df_selected)

  # Check the variance and cumulative variance by factors to see the amount of
  # variance explained by the model
  df_fa_var = pd.DataFrame(fa.get_factor_variance(),index=['Variance',
              'Proportional Var','Cumulative Var'])
  print(f'\nVariance and Cumulative Variance explained by each factor: \
  \n{df_fa_var}')

  # Create a DataFrame with the factors
  df_fa = pd.DataFrame(data = fa_components)

  # Rename columns
  df_fa.columns = [f'fa_{i}' for i in range(1, number_factors + 1)]

  # Concatenate the principal components to the original dataframe
  df_combined = pd.concat([df, df_fa], axis = 1)

  return df_combined


# Define a list of selected variables for FA
selected_variables = [
    #'age_scaled',
    'income_year_scaled',
    'wage_month_scaled',
    #'bank_accounts_scaled',
    #'interest_rate_complete_scaled',
    #'number_of_operations_scaled',
    #'account_movements_index_complete',
    'balance_complete_scaled',
    'insurance_products_scaled',
    'current_debt_scaled',
    'payment_delay_scaled',
    #'late_payments_complete_scaled',
    'deposit_complete',
]

# Setting the threshold of Eigenvalues which influence the number of factors
ei_value = 1

df_factors = perform_factor_analysis(
  df_in = df_scaled,
  variables = selected_variables,
  ei_value = ei_value
)
```

### 5.12.2 Principal component analysis <a class="anchor" id="section_5_12_2"></a>

We use principal component analysis (PCA) on continuous variables to decrease the dimensionality of the data. Due to the fact that PCA is sensitive to the magnitude of the values in the data, we use the scaled versions of the variables.

We choose 85% as the cumulative variance threshold to determine the number of principal components.

Finally, we incorporate the principal components into our dataset. During the feature selection phase, we will have the option to choose between using the original variables or the principal components for model development.

```python
def perform_principal_component_analysis(df_in, variables, variance):
  '''
    Perform principal component analysis (PCA) on a given DataFrame with
    specified variables.

    Args:
    df_in (pd.DataFrame): The input DataFrame.
    variables (list): A list of column names from df_in to be used for PCA.
    variance (float): The fraction of total variance to retain during PCA.
                      This should be a float between 0.0 and 1.0.

    Returns:
    pd.DataFrame: A DataFrame containing the original data and the
                  principal components. The principal components are named as
                  'pca_1', 'pca_2', ..., 'pca_n'.
  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Create a DataFrame with the selected variables
  df_selected = df[variables]

  # Applying PCA to retain specified amount of the total variance
  pca = PCA(n_components = variance)
  principal_components = pca.fit_transform(df_selected)

  # Get the number of selected variables
  number_of_variables = len(variables)
  print(f'Number of selected variables: {number_of_variables}')

  # Get the number of principal components
  number_of_components = pca.n_components_
  print(f'\nNumber of principal components: {number_of_components}')

  # Degree of dimensionality reduction
  dimensionality_reduction = number_of_variables - number_of_components
  print(f'\nDegree of dimensionality reduction: {dimensionality_reduction}')

  # Get the amount of variance explained by each principal component
  variance_explained = pca.explained_variance_ratio_
  var = json.dumps(list(variance_explained), indent = 4).replace('"', "'")
  print(f'\nVariance explained by each component:\n{var}')

  # Get the cumulative variance explained
  cumulative_variance_explained = variance_explained.cumsum()
  cum_var = json.dumps(list(cumulative_variance_explained),
                   indent = 4).replace('"', "'")
  print(f'\nCumulative variance explained:\n{cum_var}')

  # Create a DataFrame with the principal components
  df_pca = pd.DataFrame(data = principal_components)

  # Rename columns
  df_pca.columns = [f'pca_{i}' for i in range(1, number_of_components + 1)]

  # Concatenate the principal components to the original dataframe
  df_combined = pd.concat([df, df_pca], axis = 1)

  return df_combined


# Define a list of selected variables for PCA
selected_variables = [
    'age_scaled',
    'income_year_scaled',
    'wage_month_scaled',
    'bank_accounts_scaled',
    'interest_rate_complete_scaled',
    'number_of_operations_scaled',
    'balance_complete_scaled',
    'insurance_products_scaled',
    'current_debt_scaled',
    'payment_delay_scaled',
    'late_payments_complete_scaled'
]

# Setting amount of total variance to be explained by the principal components
total_variance = 0.85

df_components = perform_principal_component_analysis(
  df_in = df_factors,
  variables = selected_variables,
  variance = total_variance
)
```

# 6. Feature engineering <a class="anchor" id="chapter_6"></a>

In this chapter, we create new features by performing actions such as binning, arithmetic operations, or by combining two or more of them to create a combined feature. We experiment with various approaches and combinations to determine the most effective strategy for building a model, making it somewhat of a creative process.

First we add plus one to the score to ensure that all labels contain positive integer values. Some of the methods, for example XGBoost are unable handle negative integer values in the target.

After that we add a few bins variables like age, wage_month and income_year. We have also included summaries for various types of insurance, such as life, vehicle, non-life, and others. Additionally, we find some interesting ratios, like the balance-to-wage ratio and the balance-to-yearly-income ratio. Since there are many job types in the dataset, we grouped similar jobs together, such as combining features for office and technical jobs.  

We keep the original variables as well, thus providing us the flexibility to choose the most suitable features during the model development phase.

```python
def perform_feature_engineering(df_in):
  '''
    Perform feature engineering on the input DataFrame.

    Args:
    df_in (pandas.DataFrame): The input DataFrame containing the preprocessed
                              training data.

    Returns:
    pandas.DataFrame: The modified DataFrame, ready for further processing.
  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Add plus one to score to ensure that all labels contain a positive
  # integer value
  df['adjusted_score'] = df['score'] + 1

  # Round the values in the 'interest_rate_complete' column to 3 decimal places
  df['interest_rate_complete_rounded'] = df['interest_rate_complete'].round(3)
  # Convert new column to Pandas nullable integer type ('Float64')
  df['interest_rate_complete_rounded'] = (
      df['interest_rate_complete_rounded'].astype('Float64'))

  # Define the edges and lables for categorical binning
  bin_edges = [0, 29, 49, np.inf]
  bin_labels = [1, 2, 3]
  # Perform categorical binning
  df['age_bins'] = pd.cut(df['age'], bins = bin_edges, labels = bin_labels,
                          right = False)
  # Convert new column to Pandas nullable integer type ('Int64')
  df['age_bins'] = df['age_bins'].astype('Int64')

  # New feature with ratio: balance / income_year
  df['ratio_balance_income'] = df['balance_complete'] / df['income_year']

  # New feature with ratio: balance / wage_month
  df['ratio_balance_wage'] = df['balance_complete'] / df['wage_month']

  # Scaling ratio features
  df['ratio_balance_income_scaled'] = ((df['ratio_balance_income'] -
    df['ratio_balance_income'].mean(axis=0)) /
    df['ratio_balance_income'].std(axis=0))
  df['ratio_balance_wage_scaled'] = ((df['ratio_balance_wage'] -
    df['ratio_balance_wage'].mean(axis=0)) /
    df['ratio_balance_wage'].std(axis=0))

  # New feature with count of life insurances
  df['count_life_insurance'] = (
      df[['insurance_class_life','insurance_class_medical']].gt(0).sum(axis=1))

  # New features with count of non-life insurances
  df['count_vehicle_insurance'] = (
    df[['insurance_class_car','insurance_class_motorbike']].gt(0).sum(axis=1))
  df['count_house_insurance'] = (df[['insurance_class_first_house',
    'insurance_class_second_house']].gt(0).sum(axis=1))
  df['count_other_insurance'] = (df[['insurance_class_financial',
    'insurance_class_other']].gt(0).sum(axis=1))

  # New job features - unify office and technical jobs (job type blue collar,
  # artist, doctor and self_employed retain in original features)
  df['job_office_new'] = (df[['job_white_collar',
    'job_financial_advisor','job_pr_manager','job_project_manager',
    'job_journalist','job_accountant']].gt(0).sum(axis=1))
  df['job_technical_new'] = (df[['job_technician','job_civil_engineer',
    'job_developer','job_researcher']].gt(0).sum(axis=1))

  # Try a slightly different version of new job features
  df['job_finance_new'] = (df[['job_accountant','job_financial_advisor'
	]].gt(0).sum(axis=1))
  df['job_management_new'] = (df[['job_project_manager','job_pr_manager'
    ]].gt(0).sum(axis=1))
  df['job_creative_new'] = (df[['job_artist','job_journalist'
    ]].gt(0).sum(axis=1))

  # Combine the two vehicle insurance types into one feature
  df['insurance_class_vehicle'] = (df['insurance_class_car'] +
                                   df['insurance_class_motorbike'])
  df['insurance_class_vehicle'] = (
  df['insurance_class_vehicle'].astype('Int64'))

  # Combine the two property insurance types into one feature
  df['insurance_class_property'] = (df['insurance_class_first_house'] +
                                    df['insurance_class_second_house'])
  df['insurance_class_property'] = (
  df['insurance_class_property'].astype('Int64'))

  # Create a simplified version of the 'account_movements_index' with
  # (-1 = decreasing, 0 = stable, 1 = increasing)
  df['account_movements_index_simplified'] = (
  np.where(df['account_movements_index_complete'] == -2, -1,
           np.where(df['account_movements_index_complete'] == 2, 1, 0)))

  # Round the 'wage_month' values to the nearest hundred to reduce noise
  df['wage_month_rounded'] = (df['wage_month'].apply(lambda x:
    round(x / 100) * 100))
  df['wage_month_rounded'] = df['wage_month_rounded'].astype('Float64')

  # Round the 'balance_complete' values to the nearest hundred to reduce noise
  df['balance_rounded'] = (df['balance_complete'].apply(lambda x:
    round(x / 100) * 100))
  df['balance_rounded'] = df['balance_rounded'].astype('Float64')

  # Calculate the total payment delay by multiplying 'payment_delay' and
  # 'late_payments_complete' columns
  df['payment_delay_total'] = df['payment_delay'] * df['late_payments_complete']
  bin_edges = [0, 1, 30, np.inf]
  bin_lables = [0, 1, 2]
  df['payment_delay_total_bins'] = (
  pd.cut(df['payment_delay_total'],
    bins = bin_edges, labels = bin_lables, right = False))
  df['payment_delay_total_bins'] = (
  df['payment_delay_total_bins'].astype('Int64'))

  return df


df_xgb_input = perform_feature_engineering(df_preprocessed)
df_xgb_input.head()
```

### 7.2.1 Random split <a class="anchor" id="section_7_2_1"></a>

We split our dataset to 80% training set and 20% test set. In the split function we use stratify by the target variable. We set random seet and check details about the train and test set.

---

**<u>Important:</u>** The random split is used for building the model!

```python
def perform_random_train_test_split(df_in, label, test_size, random_seed):
  '''
    Split the DataFrame into training and testing sets for machine learning.

    Args:
      df_in (pandas.DataFrame): The input DataFrame containing the dataset.
      label (str): The target variable column name.
      test_size (float): The proportion of the dataset to include in the test
                        split (e.g., 0.2 for 20%).
      random_seed (int): The random seed for reproducibility.

    Returns:
      tuple: A tuple containing six DataFrames:
            X_train, X_val, X_test, y_train, y_val and y_test.
  '''

  # Extract the feature data
  X = df_in.drop(columns = ['id'] + label, axis = 1)

  # Extract the target variable column
  y = df_in[label]

  # Perform the train and test set split with stratification and a random seed
  X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    stratify = y,
                                                    test_size = (test_size),
                                                    random_state = random_seed)

  # Print the train and test set split details
  print(f'Train/test set split:')
  print(f'X_train ({len(X_train) / len(df_in) * 100}%): {X_train.shape}')
  print(f'y_train ({len(y_train) / len(df_in) * 100}%): {y_train.shape}')
  print(f'X_test ({len(X_test) / len(df_in) * 100}%): {X_test.shape}')
  print(f'y_test ({len(y_test) / len(df_in) * 100}%): {y_test.shape}')

  return X_train, X_test, y_train, y_test

# Perform train and test set split
X_train, X_test, y_train, y_test = perform_random_train_test_split(
  df_in = df_xgb_input,
  label = ['adjusted_score'],
  test_size = 0.2,
  random_seed = seed
)
```

```python
def perform_customer_train_test_split(df_in, label, month, test_size,
                                      n_folds, random_seed):
  '''
    Splits the DataFrame into training and testing sets based on customer data
    for a specific month. The function also prepares cross-validation folds
    for the training process.

    Args:
      df_in (DataFrame): The input DataFrame containing the data.
      label (list): A list containing the name of the target variable column.
      month (int): The month based on which the train-test split is performed.
      test_size (float): The proportion of the dataset to include in the test
                        split (e.g., 0.2 for 20%).
      n_folds (int): The number of folds for cross-validation.
      random_seed (int): The random seed for reproducibility.

    Returns:
      tuple: A tuple containing the following elements in order:
        - X_train (DataFrame): Features for the training set.
        - X_test (DataFrame): Features for the testing set.
        - y_train (Series/DataFrame): Target variable for the training set.
        - y_test (Series/DataFrame): Target variable for the testing set.
        - cv_folds (list): A list of tuples, each containing train and
                           validation indices for each fold.
  '''

  # Make a copy of the input DataFrame to avoid modifying the original DataFrame
  df = df_in.copy()

  # Select data for the specified month
  X = df[df['month'] == month][['id']]
  y = df[df['month'] == month][label]

  # Split the data into training and testing sets
  X_train_ids, X_test_ids, y_train, y_test = train_test_split(
      X,
      y,
      stratify = y,
      test_size = test_size,
      random_state = seed
  )

  # Construct the training and testing sets
  X_train = df[df['id'].isin(X_train_ids['id'])]
  X_test = df[df['id'].isin(X_test_ids['id'])]
  y_train = df[df['id'].isin(X_train_ids['id'])][label]
  y_test = df[df['id'].isin(X_test_ids['id'])][label]

  # Set up cross-validation using StratifiedKFold
  skf = StratifiedKFold(
      n_splits = n_folds,
      shuffle = True,
      random_state = random_seed
  )

  # Prepare data for cross-validation
  X_train_month = X_train[X_train['month'] == month]
  y_train_month = y_train[X_train['month'] == month]

  # Create indices for cross-validation folds
  fold_indices = {}
  cv_folds = []
  for fold, (train_index, val_index) in enumerate(skf.split(X_train_month,
                                                            y_train_month)):
    train_ids = X_train_month.iloc[train_index]['id']
    val_ids = X_train_month.iloc[val_index]['id']

    X_train_fold = X_train[X_train['id'].isin(train_ids)]
    X_val_fold = X_train[X_train['id'].isin(val_ids)]

    fold_indices[f"Fold_{fold+1}"] = {
        "train": X_train_fold.index.tolist(),
        "validation": X_val_fold.index.tolist()
    }

    cv_folds.append((X_train_fold.index, X_val_fold.index))

  # Reset index
  X_train = X_train.reset_index()
  y_train = y_train.reset_index()

  # Map new fold indices
  index_mapping = {value: index for index, value in enumerate(X_train['index'])}
  new_folds = []
  for tuples in cv_folds:
    temp = []
    for array in tuples:
      mapped_list = [index_mapping[i] for i in array]
      mapped_index = pd.Index(mapped_list, dtype='int64')
      temp.append(mapped_index)
    new_folds.append(tuple(temp))

  # Before modifying the DataFrames, create explicit copies
  X_train = X_train.copy()
  y_train = y_train.copy()
  X_test = X_test.copy()

  # Now its safe to perform in-place modifications
  X_train.drop(['index', 'id', 'score'], axis = 1, inplace = True)
  y_train.drop('index', axis = 1, inplace = True)
  X_test.drop(['id', 'score'], axis = 1, inplace = True)

  # Print the train and test set split details
  print(f'Train/test set split:')
  print(f'X_train ({len(X_train) / len(df_in) * 100}%): {X_train.shape}')
  print(f'y_train ({len(y_train) / len(df_in) * 100}%): {y_train.shape}')
  print(f'X_test ({len(X_test) / len(df_in) * 100}%): {X_test.shape}')
  print(f'y_test ({len(y_test) / len(df_in) * 100}%): {y_test.shape}')

  # Print details of the cross-validation folds
  print(f'\n\nCross-validation folds for training process:')
  for i in range(1, n_folds+1):
    print(f"Fold {i} (train):", len(fold_indices[f'Fold_{i}']['train']))
    print(f"Fold {i} (validation):",
          len(fold_indices[f'Fold_{i}']['validation']))


  return X_train, X_test, y_train, y_test, new_folds


# !IMPORTANT: Although this approach is currently not in use, the notebook
# can be easily adapted to use this approach instead of random split.

# Perform train and test set split
# X_train, X_test, y_train, y_test, cv_folds = (
#     perform_customer_train_test_split(
#       df_in = df_xgb_input,
#       label = ['adjusted_score'],
#       month = 1,
#       n_folds = 5,
#       test_size = 0.2,
#       random_seed = seed
#     )
# )
```

#### 7.4.1.2 Build Logistic Regression model <a class="anchor" id="section_7_4_1_2"></a>

```python
def train_lr_model(X_train, y_train, X_test, y_test, params, cv_folds,
                   random_seed, metric, weights, n_jobs = 1, verbosity = 0):
  '''
    Train a Logistic Regression classifier using GridSearchCV for
    hyperparameter tuning.

    Args:
      X_train (array-like): Training data features.
      y_train (array-like): Training data labels.
      X_test (array-like): Test data features.
      y_test (array-like): Test data labels.
      params (dict): Dictionary of Logistic Regression hyperparameters.
      cv_folds (int): Number of cross-validation folds.
      random_seed (int): Random seed for reproducibility.
      metric (str): Evaluation metric for GridSearchCV.
      weights (array-like): Sample weights for training.
      n_jobs (int): Number of CPU cores to use for parallel processing.
      verbosity (int): Verbosity level.

    Returns:
      final_model (LogisticRegression): Trained Logistic Regression classifier
                                        with the best hyperparameters.
  '''

  # Reshape the target variable
  y_train = y_train.squeeze()

  # Define the parameter grid for GridSearchCV
  param_grid = {
      'solver': params['solver'],
      'C': params['C'],
      'penalty': params['penalty'],
      'max_iter': params['max_iter'],
      'multi_class': params['multi_class']
  }

  # Initialize Logistic Regression classifier
  lr_model = LogisticRegression(
      random_state = random_seed,
      verbose = verbosity
  )

  # Setup stratified k-fold cross-validation
  stratified_cv = StratifiedKFold(
      n_splits = cv_folds,
      random_state = random_seed,
      shuffle = True
  )

  # Initialize GridSearchCV
  grid_search = GridSearchCV(
      estimator = lr_model,
      param_grid = param_grid,
      scoring = metric,
      cv = stratified_cv,
      n_jobs = n_jobs,
      verbose = verbosity
  )

  # Calculate running time
  tic = time.perf_counter()

  # Fit the model using grid search
  grid_search.fit(
      X_train,
      y_train,
      sample_weight = weights
  )

  # Retrieve the best hyperparameters from the grid search results
  best_params = grid_search.best_params_

  # Initialize the Logistic Regression classifier with the best hyperparameters
  final_model = LogisticRegression(
      C = best_params['C'],
      penalty = best_params['penalty'],
      solver = best_params['solver'],
      random_state = random_seed,
      verbose = verbosity
  )

  # Fit the final Logistic regression model on the training data
  final_model.fit(
      X_train,
      y_train,
      sample_weight = weights
  )

  # Calculate running time
  toc = time.perf_counter()

  # Evaluate the model on training data using f1 weighted
  y_pred_train = final_model.predict(X_train)
  f1_train = f1_score(y_train, y_pred_train, average = 'weighted')

  # Evaluate the model on test data using f1 weighted
  y_pred_test = final_model.predict(X_test)
  f1_test = f1_score(y_test, y_pred_test, average = 'weighted')


  # Print model performance details
  print(f'Running time: {toc - tic} seconds')
  print(f'\nBest params derived by grid search: \
        {json.dumps(grid_search.best_params_, indent = 4)}')
  print(f'\nF1 score weighted (train): {f1_train * 100.0}%')
  print(f'\nF1 score weighted (test): {f1_test * 100.0}%')
  print(f'\nProportion of overfitting: {(f1_train - f1_test) * 100.0}%')

  return final_model


# Define parameters for hyperparameter tuning
params = {
    'solver': ['saga','newton-cg','lbfgs'],
    'C': [1e-2, 1e-1, 1, 10, 100],
    'penalty': ['l2'],
    'max_iter': [10000],
    'multi_class': ['auto']
}

# Setting the number of cross-validation folds
cv_folds = 5

# Train the Logistic Regression model with the specified parameters
lr_model = train_lr_model(
    X_train = X_train[lr_features],
    y_train = y_train,
    X_test = X_test[lr_features],
    y_test = y_test,
    params = params,
    cv_folds = cv_folds,
    random_seed = seed,
    metric = 'f1_weighted',
    weights = weights
)
```

```python
from sklearn.feature_selection import SelectFromModel

def stepwise_feature_selection_lr(X_train, y_train, params, random_seed):
    '''
    Perform stepwise feature selection based on Logistic Regression model coefficients.

    Args:
      X_train (array-like): Training data features.
      y_train (array-like): Training data labels.
      params (dict): Dictionary of Logistic Regression hyperparameters.
      random_seed (int): Random seed for reproducibility.

    Returns:
      best_features (list): List of selected feature names in the best combination.
      best_score (float): Score (e.g., F1 weighted) of the best feature combination.
    '''

    # Reshape the target variable
    y_train = y_train.squeeze()

    # Initialize an empty list to store selected features
    selected_features = []

    # Get the total number of features
    total_features = X_train.shape[1]

    # Initialize Logistic Regression classifier
    lr_model = LogisticRegression(random_state=random_seed)

    # Set up Logistic Regression with specified parameters
    lr_model.set_params(**params)

    # Initialize variables to store the best feature combination and its score
    best_features = []
    best_score = -np.inf

    # Iterate through each feature
    for i in range(total_features):
        # Initialize SelectFromModel with Logistic Regression model
        selector = SelectFromModel(lr_model, threshold=-np.inf, max_features=len(selected_features) + 1)

        # Fit feature selector to training data
        selector.fit(X_train, y_train)

        # Get support mask indicating which features were selected
        selected_features_mask = selector.get_support()

        # If no new features are selected, stop the iteration
        if not np.any(selected_features_mask):
            break

        # Get selected feature names
        # Assuming X_train is a DataFrame
        selected_feature_names = X_train.columns[selected_features_mask]

        # Remove already selected features
        selected_feature_names = [feature for feature in selected_feature_names if feature not in selected_features]

        # Select the feature that improves the model the most
        best_feature = None
        best_feature_score = -np.inf
        for feature in selected_feature_names:
            # Add the feature to the selected features list
            temp_selected_features = selected_features + [feature]

            # Fit the model with the selected features
            lr_model.fit(X_train[temp_selected_features], y_train)

            # Calculate the score (e.g., F1 weighted) of the model
            score = f1_score(y_train, lr_model.predict(X_train[temp_selected_features]), average='weighted')

            # Check if the current feature improves the model
            if score > best_feature_score:
                best_feature_score = score
                best_feature = feature

        # Add the best feature to the selected features list
        selected_features.append(best_feature)

        print("Step", i+1, "- Selected Variable:", best_feature, "- Score (F1 weighted):", best_feature_score)

        # Update the best feature combination and its score if necessary
        if best_feature_score > best_score:
            best_score = best_feature_score
            best_features = selected_features.copy()

    return best_features, best_score


# Define parameters for Logistic Regression
lr_params = {
    'solver': 'saga',
    'C': 100,
    'penalty': 'l2',
    'max_iter': 10000,
    'multi_class': 'auto'
}

# Perform stepwise feature selection
best_features, best_score = stepwise_feature_selection_lr(X_train[lr_features], y_train, lr_params, random_seed=seed)

print("\nBest Feature Combination:")
print(best_features)
print("Best Score (F1 weighted):", best_score)
```

#### 7.4.2.2 Build Random Forest model <a class="anchor" id="section_7_4_2_2"></a>

```python
def train_rf_model(X_train, y_train, X_test, y_test, params, cv_folds,
                   random_seed, metric, weights, n_jobs = 1, verbosity = 0):
    '''
      Train a Random Forest classifier using GridSearchCV for hyperparameter
      tuning.

      Args:
        X_train (array-like): Training data features.
        y_train (array-like): Training data labels.
        X_test (array-like): Test data features.
        y_test (array-like): Test data labels.
        params (dict): Dictionary of Random Forest hyperparameters.
        cv_folds (int): Number of cross-validation folds.
        random_seed (int): Random seed for reproducibility.
        metric (str): Evaluation metric for GridSearchCV.
        weights (array-like): Sample weights for training.
        n_jobs (int): Number of CPU cores to use for parallel processing.
        verbosity (int): Verbosity level for Random Forest.

      Returns:
        final_model (RandomForestClassifier): Trained Random Forest classifier
                                              with the best hyperparameters.
    '''

    # Reshape the target variable
    y_train = y_train.squeeze()

    # Define the parameter grid for GridSearchCV
    param_grid = {
        'n_estimators': params['n_estimators'],
        'max_depth': params['max_depth'],
        'min_samples_split': params['min_samples_split'],
        'min_samples_leaf': params['min_samples_leaf'],
        'max_features': params['max_features'],
        'bootstrap': params['bootstrap'],
        'criterion': params['criterion']
    }

    # Initialize Random Forest classifier
    rf_model = RandomForestClassifier(
        random_state = random_seed,
        n_jobs = n_jobs,
        verbose = verbosity
    )

    # Setup stratified k-fold cross-validation
    stratified_cv = StratifiedKFold(
        n_splits = cv_folds,
        random_state = random_seed,
        shuffle = True
    )

    # Initialize GridSearchCV
    grid_search = GridSearchCV(
        estimator = rf_model,
        param_grid = param_grid,
        scoring = metric,
        cv = stratified_cv,
        n_jobs = n_jobs,
        verbose = verbosity
    )

    # Calculate running time
    tic = time.perf_counter()

    # Fit the model using grid search
    grid_search.fit(X_train, y_train, sample_weight = weights)

    # Retrieve the best hyperparameters from the grid search results
    best_params = grid_search.best_params_

    # Initialize the Random Forest classifier with the best hyperparameters
    final_model = RandomForestClassifier(
        n_estimators = best_params['n_estimators'],
        max_depth = best_params['max_depth'],
        min_samples_split = best_params['min_samples_split'],
        min_samples_leaf = best_params['min_samples_leaf'],
        max_features = best_params['max_features'],
        bootstrap = best_params['bootstrap'],
        criterion = best_params['criterion'],
        random_state = random_seed,
        n_jobs = n_jobs,
        verbose = verbosity
    )

    # Fit the final Random Forest model on the training data
    final_model.fit(X_train, y_train, sample_weight = weights)

    # Calculate running time
    toc = time.perf_counter()

    # Evaluate the model on training data using f1 weighted
    y_pred_train = final_model.predict(X_train)
    f1_train = f1_score(y_train, y_pred_train, average = 'weighted')

    # Evaluate the model on test data using f1 weighted
    y_pred_test = final_model.predict(X_test)
    f1_test = f1_score(y_test, y_pred_test, average = 'weighted')

    # Print model performance details
    print(f'Running time: {toc - tic} seconds')
    print(f'\nBest params derived by grid search: \
         {json.dumps(grid_search.best_params_, indent = 4)}')
    print(f'\nF1 score weighted (train): {f1_train * 100.0}%')
    print(f'\nF1 score weighted (test): {f1_test * 100.0}%')
    print(f'\nProportion of overfitting: {(f1_train - f1_test) * 100.0}%')

    return final_model

# Define parameters for hyperparameter tuning
params = {
    'n_estimators': [100],
    'max_depth': [20],
    'min_samples_split': [5,10],
    'min_samples_leaf': [4],
    'max_features': ['sqrt'],
    'bootstrap': [False],
    'criterion': ['gini']
}

# Setting the number of cross-validation folds
cv_folds = 5

# Train the Random Forest model with the specified parameters
rf_model = train_rf_model(
    X_train = X_train[rf_features],
    y_train = y_train,
    X_test = X_test[rf_features],
    y_test = y_test,
    params = params,
    cv_folds = cv_folds,
    random_seed = seed,
    metric = 'f1_weighted',
    weights = weights
)
```

### 7.4.3 XGBoost <a class="anchor" id="section_7_4_3"></a>

XGBoost is one of the most popular machine learning frameworks among data scientists. #### 7.4.3.2 Build XGBoost model <a class="anchor" id="section_7_4_3_2"></a>

```python
def train_xgb_model(X_train, y_train, X_test, y_test, num_class, params,
                    cv_folds, random_seed, metric, weights, device = 'cpu',
                    early_stopping_rounds = 10, n_jobs = 1, verbosity = 0):
  '''
    Train an XGBoost classifier using GridSearchCV for hyperparameter tuning.

    Args:
      X_train (array-like): Training data features.
      y_train (array-like): Training data labels.
      X_test (array-like): Test data features.
      y_test (array-like): Test data labels.
      num_class (int): Number of classes in the classification problem.
      params (dict): Dictionary of XGBoost hyperparameters.
      cv_folds (int): Number of cross-validation folds.
      random_seed (int): Random seed for reproducibility.
      metric (str): Evaluation metric for GridSearchCV.
      weights (array-like): Sample weights for training.
      device (str): Device to use for XGBoost ('cpu' or 'cuda').
      early_stopping_rounds (int): Number of rounds to perform early stopping.
      n_jobs (int): Number of CPU cores to use for parallel processing.
      verbosity (int): Verbosity level for XGBoost.

    Returns:
      final_model (XGBClassifier): Trained XGBoost classifier with the best
                                   hyperparameters.
  '''

  # Define the parameter grid for GridSearchCV
  param_grid = {
      'num_class': [num_class],
      'objective': params['objective'],
      'eta': params['eta'],
      'n_estimators': params['n_estimators'],
      'max_depth': params['max_depth'],
      'min_child_weight': params['min_child_weight'],
      'subsample': params['subsample'],
      'colsample_bytree': params['colsample_bytree'],
      'gamma': params['gamma'],
      'reg_alpha': params['reg_alpha'],
      'reg_lambda': params['reg_lambda'],
      'device' : [device]
  }

  # Initialize XGBoost classifier
  xgb_model = xgb.XGBClassifier(
      device = ('gpu' if device == 'cuda' else 'cpu'),
      seed = random_seed,
      verbosity = verbosity
  )

  # Setup stratified k-fold cross-validation
  stratified_cv = StratifiedKFold(
      n_splits = cv_folds,
      random_state = random_seed,
      shuffle = True
  )

  # Initialize GridSearchCV
  grid_search = GridSearchCV(
      estimator = xgb_model,
      param_grid = param_grid,
      scoring = metric,
      cv = stratified_cv,
      n_jobs = n_jobs,
      verbose = verbosity
  )

  # Calculate running time
  tic = time.perf_counter()

  # Fit the model using grid search
  grid_search.fit(
      X_train,
      y_train,
      sample_weight = weights,
      verbose = verbosity
  )

  # Retrieve the best number of estimators from the grid search results
  n_estimators = grid_search.best_params_['n_estimators']

  # Retrieve the best hyperparameters from the grid search results
  best_params = dict((key, value) for key, value in \
                     grid_search.best_params_.items() \
                     if key not in ['n_estimators'])


  # Create a DMatrix for XGBoost using the training data
  dtrain = xgb.DMatrix(X_train, label = y_train)

  # Perform cross-validation for hyperparameter tuning
  cv_results = xgb.cv(
      dtrain = dtrain,
      params = best_params,
      num_boost_round = n_estimators,
      nfold = cv_folds,
      metrics = {'mlogloss'},
      early_stopping_rounds = early_stopping_rounds,
      seed = random_seed
  )

  # Determine the best number of boosting rounds based on cv results
  n_estimators_best = cv_results.shape[0]

  # Initialize the XGBoost classifier with the best hyperparameters
  final_model = xgb.XGBClassifier(
      objective = best_params['objective'],
      eta = best_params['eta'],
      n_estimators = n_estimators_best,
      max_depth = best_params['max_depth'],
      min_child_weight = best_params['min_child_weight'],
      subsample = best_params['subsample'],
      colsample_bytree = best_params['colsample_bytree'],
      gamma = best_params['gamma'],
      reg_alpha = best_params['reg_alpha'],
      reg_lambda = best_params['reg_lambda'],
      num_class = num_class,
      device = ('gpu' if device == 'cuda' else 'cpu'),
      seed = random_seed,
      verbosity = verbosity
  )

  # Fit the final XGBoost model on the training data
  final_model.fit(
      X_train,
      y_train,
      sample_weight = weights,
      verbose = verbosity
  )

  # Calculate running time
  toc = time.perf_counter()

  # Evaluate the model on training data using f1 weighted
  y_pred_train = final_model.predict(X_train)
  f1_train = f1_score(y_train, y_pred_train, average = 'weighted')

  # Evaluate the model on test data using f1 weighted
  y_pred_test = final_model.predict(X_test)
  f1_test = f1_score(y_test, y_pred_test, average = 'weighted')


  # Print model performance details
  print(f'Running time: {toc - tic} seconds')
  print(f'Number of boosting rounds for early stopping: {n_estimators_best}')
  print(f'\nBest params derived by grid search: \
        {json.dumps(grid_search.best_params_, indent = 4)}')
  print(f'\nF1 score weighted (train): {f1_train * 100.0}%')
  print(f'\nF1 score weighted (test): {f1_test * 100.0}%')
  print(f'\nProportion of overfitting: {(f1_train - f1_test) * 100.0}%')

  return final_model


# Define parameters for hyperparameter tuning
params = {
    'objective': ['multi:softprob'],
    'eta': [0.165],
    'n_estimators': [1000],
    'max_depth': [15],
    'min_child_weight': [12],
    'subsample': [0.65],
    'colsample_bytree': [0.9],
    'gamma': [0.8],
    'reg_alpha': [0.3],
    'reg_lambda': [0.1]
}

# Setting the number of cross-validation folds
cv_folds = 5

# Train the XGBoost model with the specified parameters
xgb_model = train_xgb_model(
    X_train = X_train[xgb_features],
    y_train = y_train,
    X_test = X_test[xgb_features],
    y_test = y_test,
    num_class = 3,
    params = params,
    cv_folds = cv_folds,
    random_seed = seed,
    metric = 'f1_weighted',
    weights = weights,
    device = 'cuda'
)
```

### 7.5.4 Predict probabilities and use custom threshold for class assignment <a class="anchor" id="section_7_5_4"></a>

In this chapter we predict probabilities using our final model and compute confusion matrices for default and with custom thresholds. Here we focus on the Class 2 (score 1). As an example we use a custom threshold of 0.25 for the class assignment.

In real-world applications, the costs of false negatives and false positives often differ significantly. By setting a custom threshold, we can manage this aspect more effectively according to a particular use case. For example, we have the option to lower the occurrence of false negatives, even if it results in an increase in the rate of false positives. In certain scenarios, such an approach may be preferable and can enhance the overall outcome.

```python
# Make predictions using the XGBoost model
y_pred_test = xgb_model_final.predict(X_test[xgb_features_reduced_final])

# Set a custom threshold for class assignment
label = 2
custom_threshold = 0.25

# Get class probabilities for each class
probabilities = xgb_model_final.predict_proba(X_test[xgb_features_reduced_final])

# Apply custom threshold to assign class labels
y_pred_test_cut = np.where(
    probabilities[:, label] >= custom_threshold, 2,
    np.where(probabilities[:, 0] >= custom_threshold, 0, 1)
)

# Compute confusion matrices for default and custom thresholds
cm_default = confusion_matrix(y_test, y_pred_test)
cm_custom_threshold = confusion_matrix(y_test, y_pred_test_cut)

# Create a figure with two subplots for confusion matrices
fig, axes = plt.subplots(1, 2, figsize = (16, 6))

# Plot the confusion matrix with default threshold
sns.heatmap(cm_default, annot = True, fmt = 'd', cmap = 'Blues', ax = axes[0])
axes[0].set_title('Confusion matrix (default)')
axes[0].set_ylabel('True label')
axes[0].set_xlabel('Predicted label')

# Plot the confusion matrix with custom threshold
sns.heatmap(cm_custom_threshold, annot = True, fmt = 'd', cmap = 'Blues',
            ax = axes[1])
axes[1].set_title(f'Confusion Matrix (custom threshold = {custom_threshold})')
axes[1].set_ylabel('True label')
axes[1].set_xlabel('Predicted label')

# Ensure proper layout and display the plots
plt.tight_layout()
plt.show()
```

Define a function that can prepare the test set with the same trasformations that were applied to the training set during the data preprocessing.

```python
def prepare_input_dataset_for_model(df_in, encoder_path, is_train):

  # Only for the training set, correct invalid formats of some colums
  if is_train:
    _df_train_transformed = transform_train_data( df_in = df_in)
  else:
    _df_train_transformed = df_in.copy()
    # create the column "score" in the test set and then drop it at the end of the method
    # to ensure that all the transformations applied to the training works also with the test set
    _df_train_transformed['score']= -999

  # Clean training data
  _df_train_cleaned = clean_train_data(df_in = _df_train_transformed)

  # Decompose column insurance_class
  _df_train_decomposed = decompose_categorical_column(
      df_in = _df_train_cleaned,
      column_name = 'insurance_class',
      encoder_path = encoder_path,
      is_train = is_train
  )

  # Add column "income_year_log"
  _df_income_year_log = add_income_year_log(_df_train_decomposed)

  # Add column "wage_month_log"
  _df_wage_month_log = add_wage_month_log(_df_income_year_log)

  # Add column "wage_month_bins"
  _df_wage_month_bins = add_wage_month_bins(_df_wage_month_log)

  # Add column "interest_rate_complete"
  _df_interest_rate_complete = add_interest_rate_complete(_df_wage_month_bins)

  # Add column "account_movements_index_complete"
  _df_account_movements_index_complete = add_account_movements_index_complete(_df_interest_rate_complete)

  # Add column "balance_complete"
  _df_balance_complete = add_balance_complete(_df_account_movements_index_complete)

  # Add column "balance_cbrt"
  _df_balance_cbrt = add_balance_cbrt(_df_balance_complete)

  # Add column 'has_current_debt'
  _df_has_current_debt = add_has_current_debt(_df_balance_cbrt)

  # Add column "current_debt_bins"
  _df_current_debt_bins = add_current_debt_bins(_df_has_current_debt)

  # Add column "payment_delay_bins"
  _df_payment_delay_bins = add_payment_delay_bins(_df_current_debt_bins)

  # Add column "add_late_payments_complete"
  _df_late_payments_complete = add_late_payments_complete(_df_payment_delay_bins)

  #Add columns "has_late_payments"
  _df_has_late_payments = add_has_late_payments(_df_late_payments_complete)

  # Add column "deposit_complete"
  _df_deposit_complete = add_deposit_complete(_df_has_late_payments)

  # Apply encoding to job column
  _df_encoded = decompose_categorical_column(
      df_in = _df_deposit_complete,
      column_name = 'job',
      encoder_path = encoder_path,
      is_train = is_train,
      drop=False
  )

  # Determine a list of variables for standard scaling
  variables_to_scale = [
      'age',
      'income_year',
      'income_year_log',
      'wage_month',
      'wage_month_log',
      'bank_accounts',
      'interest_rate_complete',
      'number_of_operations',
      'balance_complete',
      'balance_cbrt',
      'insurance_products',
      'current_debt',
      'payment_delay',
      'late_payments_complete'
  ]
  _df_scaled = perform_standard_scaling(
      df_in = _df_encoded ,
      variables = variables_to_scale
  )

  # Define a list of selected variables for FA: not required
  # Define a list of selected variables for PCA: not required
  # Reorder the columns of the final preprocessed DataFrame: not required

  # Feature engineering
  _df_xgb_input = perform_feature_engineering(_df_scaled)

  if not is_train:
    # drop column score for the test set at the end of the transformations
    _df_xgb_input.drop('score',axis=1,inplace=True)

  return _df_xgb_input
```
